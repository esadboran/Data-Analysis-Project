# 1.Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<ul>
    <li>Country or region: Name of countries</li>
    <li>Overall rank: Rank of countries based on the Happiness Score</li>
    <li>Score: A metric measured in 2019 by asking the sampled people the question: "How would you rate your happiness on a scale of 0 to 10 where 10 is the happiest."</li>
    <li>GDP per capita: The extent to which GDP contributes to the calculation of the Happiness Score.</li>
    <li>Social support: The extent to which Social support contributes to the calculation of the Happiness Score</li>
    <li>Healthy life expectancy: The extent to which Life expectancy contributed to the calculation of the Happiness Score</li>
    <li>Freedom to make life choices: The extent to which Freedom contributed to the calculation of the Happiness Score</li>
    <li>Generosity: The extent to which Generosity contributed to the calculation of the Happiness Score</li>
    <li>Perceptions of corruption: The extent to which Perception of Corruption contributes to Happiness Score.</li>

    
 </ul>



# 2. Data Definition

```python
#Read CSV files 
df_2015 = pd.read_csv('World Happiness Dataset/2015.csv')
df_2016 = pd.read_csv('World Happiness Dataset/2016.csv')
df_2017 = pd.read_csv('World Happiness Dataset/2017.csv')
df_2018 = pd.read_csv('World Happiness Dataset/2018.csv')
df_2019 = pd.read_csv('World Happiness Dataset/2019.csv')
```

```python
df_2015 = df_2015.rename(columns={"Happiness Score" : "Score"})
df_2015.head()
```

```python
df_2016 = df_2016.rename(columns={"Happiness Score" : "Score"})
df_2016.head()
```

```python
df_2017 = df_2017.rename(columns={"Happiness.Score" : "Score"})
df_2017.head()
```

```python
df_2018 = df_2018.rename(columns={"Country or region" : "Country"})
df_2018.head()
```

```python
df_2019 = df_2019.rename(columns={"Country or region" : "Country"})
df_2019.head()
```

Bir korelasyon matrisi, bir veri kümesindeki değişkenler arasındaki ilişkileri gösterir. Her bir değişkenin diğer değişkenlerle olan korelasyonunu gösteren bir matristir. Matrisin diagonal elemanları her zaman 1 olur ve bu, bir değişkenin kendisiyle olan korelasyonunu gösterir. Diagonal dışındaki diğer elemanlar, iki değişken arasındaki korelasyonu gösterir. 

```python
corr_matrix = df_2015.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

# 4.Mutluluk oranları

Hangi ülkeler veya bölgeler genel mutlulukta ve mutluluğa katkıda bulunan altı faktörün her birinde en yüksek sırada yer alıyor?


2015-2016 ile 2016-2017 raporları arasında ülke sıralamaları veya puanları nasıl değişti?

```python
count=2015

for year in [df_2015,df_2016,df_2017,df_2018,df_2019]:
    
    plt.figure(figsize=(22, 6))
    
    plt.subplot(1, 4, 1)
    plt.title(f"{count} - Top 10 Happy Country")
    sns.barplot(data=year.head(10), x="Country", y="Score")
    plt.grid(False)
    plt.xticks(rotation=60)
    plt.ylabel(None)
    plt.xlabel(None)
    
    plt.subplot(1, 4, 2)
    plt.title(f"{count} - 10 unhappiest countries")
    sns.barplot(data= year.iloc[::-1].head(10), x="Country", y="Score")
    plt.grid(False)
    plt.xticks(rotation=60)
    plt.ylabel(None)
    plt.xlabel(None)
    
    plt.show()
    count += 1
```

# 4 Veri Birleşimi

```python
df = df_2015[["Country","Score"]]
df2 = df_2016[["Country","Score"]]
df3 = df_2017[["Country","Score"]]
df4 = df_2018[["Country","Score"]]
df5 = df_2019[["Country","Score"]]
```

```python
df_list = [df, df2, df3, df4, df5]

result = df_list[0]
count=2015
for df in df_list[1:]:
    result = pd.merge(result, df, on="Country")
    
result.columns = ["Country","2015","2016","2017","2018","2019"] 
result
```
